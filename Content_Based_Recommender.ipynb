## TF-IDF Text Vectorization and Transformation

Text cleaning

```python
df['detailed_description'] = [i.replace('\n', ' ') for i in df['detailed_description']]
```

```python
# Remove html tags, new line, and carriage return in description
import re
def remove_tag(html_text):
  re_html = re.compile(r'<[^>]+>||\r|\n|\t|http\S+')
  return re_html.sub(' ',html_text)
```

```python
df['detailed_description'] = df['detailed_description'].apply(remove_tag)
```

```python
df['detailed_description'] = [i.replace("'", '') for i in df['detailed_description']]
```

```python
df['detailed_description'] = [i.replace(",", '') for i in df['detailed_description']]
```

```python
import string 
df['detailed_description'] = [i.lower().translate(str.maketrans(string.punctuation, ' '*len(string.punctuation))) for i in df['detailed_description']]
```

```python
#Import TfIdfVectorizer from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

#Define a TF-IDF Vectorizer Object. Remove all english stop words such as 'the', 'a'
tfidf = TfidfVectorizer(stop_words='english')

#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(df['detailed_description'])

#Output the shape of tfidf_matrix
tfidf_matrix.shape
```

calculating the dot product will directly give us the cosine similarity score. Therefore, sklearn's linear_kernel() instead of cosine_similarities() since it is faster.

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

## Cosine Similarities 

```python
# Cosine Similarities Function
def get_recommendations(name, cosine_sim=cosine_sim):
    # index of the game 
    idx = indices[name]
    
    # similarity scores 
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar game
    sim_scores = sim_scores[1:11]
    print(sim_scores)
    # Get the recommended game indices
    game_indices = [i[0] for i in sim_scores]
    
    # Return the top 10 most similar games
    return df.loc[game_indices]
```

## Evaluation Metrics

To evaluate the recommender system:

1. To conduct survey of 30 person to rate (1-5) 5 recommendations with picture and description between 2 recommenders, then using t test to evaluate.

```python
# Download a file based on its file ID.
file_id = '19Ko3NQvlkwARTS_7uwP4AkS2n2VMGx7s'
downloaded = drive.CreateFile({'id': file_id})

# Save file in Colab memory
downloaded.GetContentFile('Responses.csv')  
```

```python
dfe = pd.read_csv('Responses.csv')
dfe = pd.DataFrame(dfe)
dfe = dfe[["Ratings 1","Ratings 2"]]
dfe.head()
```

```python
print('No. of Samples:', dfe.shape[0])
```

```python
import scipy.stats as stats

data_group1 = dfe['Ratings 1']
data_group2 = dfe['Ratings 2']

print(np.var(data_group1), np.var(data_group2))
```

```python
stats.ttest_ind(a=data_group1, b=data_group2, equal_var=True)
```

The p-value (0.344) is greater than alpha (0.05), thus, the null hypothesis of the test is not rejected. We do not have sufficient evidence to say that the ratings between the recommender systems are different.
